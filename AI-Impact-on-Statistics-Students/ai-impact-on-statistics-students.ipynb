## UNIVARIATE ANALYSIS

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gender Distribution

df['Gender'].value_counts().plot(kind='bar', title='Gender Distribution', color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# University Distribution

df['University'].value_counts().plot(kind='bar', title='University Distribution', color='lightgreen')
plt.xlabel('University')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# AI and Automation Knowledge Level

df['AI and Automation Knowledge Level'].value_counts().plot(kind='bar',title='AI Knowledge Level', color= 'orange')
plt.xlabel("Knowledge Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Concern Level

df['Concern Level'].value_counts().plot(kind='bar', title='Concern Level About AI', color='tomato')
plt.xlabel("Concern Level")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
df['Impact of AI on Job Opportunities'].value_counts().plot(kind='bar', title='Impact of AI on Job Opportunities', color='yellow')
plt.xlabel("Impact of AI on Job Opportunities")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
df['Education Prepares for AI Field'].value_counts().plot(kind='bar', title='Education Prepares for AI Field', color='red')
plt.xlabel("Education Prepares for AI Field")
plt.ylabel("Counts")
plt.show()
```

```python
# SUMMARY STATISTICS
# Distribution of AI Knowledge

df['AI and Automation Knowledge Level'].value_counts(normalize=True).plot(kind='bar', title='AI Knowledge Levels')
```

```python
# Anxiety levels by University

sns.countplot(data=df, x='University', hue='Anxiety Level')
plt.title('Anxiety Level by University')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Internship Experience vs Concern

plt.figure(figsize=(10,6))
pd.crosstab(df['Internship Experience'], df['Concern Level'], normalize='index').plot(kind='bar', stacked=True)
plt.title("Concern Level by Internship Experience")
plt.legend(title='Concern Level', bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
# GROUPED INSIGHTS
# GPA vs AI Sentiment

pd.crosstab(df['GPA'], df['Impact of AI on Job Opportunities'],normalize='index').plot(kind='bar', stacked=True)
plt.title('Perceived AI Job Impact by GPA')
plt.legend(title='Concern Level', bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
# Engineering Resilience Belief

sns.countplot(data=df, x='If I Studied Engineering, I Would Feel More Resilient to AI Impacts', hue='University')
plt.title("Engineering Resilience Belief Across Universities")
plt.xticks(rotation=45)
```

```python
# Correlation check

pd.crosstab(df["Anxiety Level"], df['Impact of AI on Career Plans'], normalize='index').plot(kind='bar', stacked=True)
plt.title("AI Anxiety vs Career Plan Impact")
plt.legend(title='Concern Level', bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
# Heatmap: University x AI Knowledge vs Concern Level

pivot = pd.crosstab(
    index=[df['University'], df['AI and Automation Knowledge Level']], 
    columns=df['Concern Level']
)

plt.figure(figsize=(12, 6))
sns.heatmap(pivot, annot=True, fmt='d', linewidths=0.5)

plt.title('Concern Level by University and AI Knowledge')
plt.ylabel('University & AI Knowledge')
plt.xlabel('Concern Level')
plt.tight_layout()
plt.show()
```

```python
# Satisfaction with Studying Statistics

satisfaction_counts = df['Satisfaction with Studying Statistics'].value_counts()

plt.figure(figsize=(8, 8))
wedges,texts, autotexts = plt.pie(
    satisfaction_counts,
    labels=None,
    autopct='%1.1f%%',
)

plt.legend(
    wedges,
    satisfaction_counts.index,
    title='Satisfaction Level',
    loc='center left',
    bbox_to_anchor=(1, 0.5)
)

plt.title('Satisfaction with Studying Statistics')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
